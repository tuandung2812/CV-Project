## Faster RCNN with no pretrained weights

## Faster RCNN with pretrained weights

```python
gt_path = './ground_truth/'
faster_rcnn_pred_path = './predictions/faster_rcnn'
faster_rcnn_pr_curves, faster_rcnn_aps = evaluation(faster_rcnn_pred_path, gt_path)
```

## Plot Precision Recall Curve

```python
fig, axs = plt.subplots(1,3)
fig.suptitle('Precision-Recall Curve', fontsize = 30, y = 1.1, fontweight = 'bold')
fig.set_figheight(8)
fig.set_figwidth(30)

plt.subplots_adjust(wspace = 0.7)

mode = ['Easy', 'Medium', 'Hard']
for i in range(3):
    axs[i].plot(faster_rcnn_pr_curves[i][:,1], faster_rcnn_pr_curves[i][:,0], label="Faster RCNN - {}".format(round(faster_rcnn_aps[i],3)),  linewidth=1.5)
    axs[i].plot(ssd_pr_curves[i][:,1], ssd_pr_curves[i][:,0], label="SSD - {}".format(round(ssd_aps[i],3)), linewidth=1.5)
    axs[i].plot(hog_svm_pr_curves[i][:,1], hog_svm_pr_curves[i][:,0], label="HOG + SVM - {}".format(round(hog_svm_aps[i],3)),  linewidth=1.5)

    axs[i].grid()
    axs[i].legend(loc="lower right",  bbox_to_anchor=(1.5,0))
    axs[i].set_xticks(np.arange(0,1.1,0.1))
    axs[i].set_yticks(np.arange(0,1.1,0.1))
    axs[i].set_title("{} set".format(mode[i]), fontsize = 15, y = 1.05)
    axs[i].set_xlabel('Recall', fontsize=13)
    axs[i].set_ylabel('Precision', fontsize=13)



```
