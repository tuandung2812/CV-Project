```python
import pandas as pd
import json
```

```python
train_df = pd.read_csv("../dataset/processed/train.csv", sep = ",")
val_df = pd.read_csv("../dataset/processed/validation.csv", sep = ",")
test_df = pd.read_csv("../dataset/processed/test.csv", sep = ",")
```

```python
train_df
```

```python
train_json = {}
train_images = train_df['name'].unique()
for img in train_images:
    train_json[img] = []

for image in train_json.keys():
    boxes = list (train_df[train_df['name'] == image]['bbox'] )
    final_boxes = []
    for box in boxes:
        box = box.split(" ")
        x_min, y_min, x_max, y_max = int(box[0]), int(box[1]), int(box[0]) + int(box[2]),int(box[1])+ int(box[3])
        final_boxes.append([x_min,y_min,x_max,y_max])
    train_json[image] = final_boxes
    
train_json
```

```python
val_json = {}
val_images = val_df['name'].unique()
for img in val_images:
    val_json[img] = []

for image in val_json.keys():
    boxes = list (val_df[val_df['name'] == image]['bbox'] )
    final_boxes = []
    for box in boxes:
        box = box.split(" ")
        x_min, y_min, x_max, y_max = int(box[0]), int(box[1]), int(box[2]), int(box[3])
        final_boxes.append([x_min,y_min,x_max,y_max])
    val_json[image] = final_boxes
    
val_json
```

```python
test_json = {}
test_images = test_df['name'].unique()
for img in test_images:
    test_json[img] = []

for image in test_json.keys():
    boxes = list (test_df[test_df['name'] == image]['bbox'] )
    final_boxes = []
    for box in boxes:
        box = box.split(" ")
        x_min, y_min, x_max, y_max = int(box[0]), int(box[1]), int(box[2]), int(box[3])
        final_boxes.append([x_min,y_min,x_max,y_max])
    test_json[image] = final_boxes
    
test_json
```

```python
with open('../dataset/processed/train.json', 'w') as f:
    json.dump(train_json, f)

with open('../dataset/processed/validation.json', 'w') as f:
    json.dump(val_json, f)

with open('../dataset/processed/test.json', 'w') as f:
    json.dump(test_json, f)
```

```python
with open('../dataset/processed/train.json', 'r') as f:
    train_json = json.load(f)

with open('../dataset/processed/validation.json', 'r') as f:
    val_json = json.load(f)


with open('../dataset/processed/test.json', 'r') as f:
    test_json = json.load(f)
```
